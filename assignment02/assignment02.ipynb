# Import packages for plotting graphs and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define my function: 

$$f(x) = \frac{\sin\left(\left(4+\frac{1}{2}\right)x\right)}{sin\left(\frac{x}{2}\right)}$$

```python
def func(x):
    y = np.sin(9*(x/2)) / np.sin(x/2)
    return y
```

# Define derivative for function :
$$f^{\prime}(x) = \frac{4\sin\left(5x\right)-5\sin\left(4x\right)}{2\sin\left(\frac{x}{2}\right)^2}$$

```python
def der_func(x):
    y = (4*np.sin(5*x) - 5*np.sin(4*x)) / (2*(np.sin(x/2))**2)
    return y
```

# Define first-order taylor approximation at $z$ :

Since first-order taylor approximation form at $z$ is $\hat{f}(x) = f(z)+\frac{\partial f}{\partial x}\left(z\right)\left(x-z\right)$, 
$$\hat{f}(x) = \frac{\sin\left(\left(4+\frac{1}{2}\right)x\right)}{sin\left(\frac{x}{2}\right)}+\frac{4\sin\left(5x\right)-5\sin\left(4x\right)}{2\sin\left(\frac{x}{2}\right)^2}\left(x-z\right)$$

```python
def taylor(x, z):
    y = func(z) + der_func(z)*(x-z)
    return y
```

# Plot the line (taylor approximation) for $\left[z-0.5, z+0.5\right]$ :

```python
def plot_taylor(z):
    x = np.arange(z-0.5, z+0.5, 0.001)
    y_taylor = taylor(x, z)
    plt.plot(x, y_taylor)
```

# Plot the graphs for the function and its derivative :

```python
x = np.arange(-3,3,0.001)
y = func(x)
der_y = der_func(x)
    
plt.plot(x, y, label='original function')
plt.plot(x, der_y, color='red', label='derivative function')
plt.legend()
plt.show()
```

# Draw lines :

```python
plt.plot(x, y)

plot_taylor(-1)
plot_taylor(-1/2)
plot_taylor(2)
```
